6. buatlah matrix (ukurannya bebas) dengan numpy dan hitunglah determinan dan invers dari matrix tersebut

```python
import numpy as np

matrix = np.matrix([[1,3,9],[7,5,6],[2,8,9]])

print("Determinan = ");np.linalg.det(matrix)
```

```python
print("Invers = ");np.linalg.inv(matrix)
```

7. buatlah 3 persamaan linear dengan 3 variabel ( dengan bentuk umum ax + by + cz = d) dan carilah nilai x,y,z tersebut

```python
import numpy as np

a = np.array([[1,3,5],[12,3,-2],[-13,21,19]])
b = np.array([13,31,2])
x,y,z = np.linalg.solve(a,b)
print("Nilai X = ",x,"\nNilai Y =",y,"\nNilai Z = ",z)
```
